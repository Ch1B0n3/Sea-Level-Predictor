```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
# Use Pandas to import the data from epa-sea-level.csv
df = pd.read_csv('/Users/chichi/Desktop/python/challenge_5/epa-sea-level.csv')
df
```

Use matplotlib to create a scatter plot using the Year column as the x-axis and the CSIRO Adjusted Sea Level column as the y-axis.

```python
fig, ax = plt.subplots(figsize=(32,10),dpi=300)
plt.scatter(df['Year'],df['CSIRO Adjusted Sea Level'])
```

Use the linregress function from scipy.stats to get the slope and y-intercept of the line of best fit. Plot the line of best fit over the top of the scatter plot. Make the line go through the year 2050 to predict the sea level rise in 2050.

```python
fig, ax = plt.subplots(figsize=(30,10),dpi=300)
plt.scatter(df['Year'],df['CSIRO Adjusted Sea Level'])

# Create first line of best fit

line_best = linregress (df['Year'],df['CSIRO Adjusted Sea Level'])
x_best = np.arange(df['Year'].min(),2050,1)
y_best = x_best * line_best.slope + line_best.intercept
plt.plot(x_best,y_best,color = 'red')

# Create second line of best fit

df_2000 = df[df['Year'] >= 2000]
line_2best = linregress (df_2000['Year'],df_2000['CSIRO Adjusted Sea Level'])
x_2best = np.arange(2000,2050,1)
y_2best = x_2best * line_2best.slope + line_2best.intercept
plt.plot(x_2best,y_2best,color = 'blue',linewidth = 5, linestyle = 'dashed')
plt.xlabel('Year',fontsize= 17)
plt.ylabel('Sea Level (inches)',fontsize= 17)
plt.title('Rise in Sea Level',fontsize= 20, weight = 'bold')

# Save the image
fig.savefig('sea_level_plot.png')
```
